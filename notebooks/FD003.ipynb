```python

# 1 -- Environment & imports
import os
import glob
import math
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping

# 2 -- data directory & quick file listing
DATA_DIR = "/home/tejasri/Desktop/springboard"
txt_files = sorted(glob.glob(os.path.join(DATA_DIR, "*.txt")))
print("Files found:", txt_files)
print()

# 3 -- helper to read CMAPSS formatted files
def read_cmapss(file_path):
    col_names = ["unit", "cycle", "op_setting_1", "op_setting_2", "op_setting_3"]
    for i in range(1, 22):
        col_names.append(f"s{i}")
    df = pd.read_csv(file_path, sep=r"\s+", header=None, names=col_names)
    return df

# 4 -- add RUL to training dataframe
def add_rul_train(df):
    df = df.copy()
    max_cycle = df.groupby("unit")["cycle"].max().reset_index()
    max_cycle.columns = ["unit", "max_cycle"]
    df = df.merge(max_cycle, on="unit", how="left")
    df["RUL"] = df["max_cycle"] - df["cycle"]
    df.drop(columns=["max_cycle"], inplace=True)
    return df

# 5 -- locate train file for FD003 and read
train_fp = os.path.join(DATA_DIR, "train_FD003.txt")
if not os.path.exists(train_fp):
    raise FileNotFoundError(f"train_FD003.txt not found in {DATA_DIR}")

train_df = read_cmapss(train_fp)
train_df = add_rul_train(train_df)
print("Train df shape:", train_df.shape)
print("Units:", train_df["unit"].nunique())
print()

# detect sensor columns
sensor_cols = [c for c in train_df.columns if c.startswith("s") or c.startswith("op_setting")]
print("Sensor cols (first 8):", sensor_cols[:8])

# 6 -- preprocessing function
def preprocess_df(df, cap_rul=125, sensor_cols=None, scaler=None, fit_scaler=True):
    df = df.copy()
    if sensor_cols is None:
        sensor_cols = [c for c in df.columns if c.startswith("s") or c.startswith("op_setting")]
    df["RUL_capped"] = df["RUL"].clip(upper=cap_rul)
    if fit_scaler:
        scaler = StandardScaler()
        scaler.fit(df[sensor_cols])
    df[sensor_cols] = scaler.transform(df[sensor_cols])
    return df, scaler, sensor_cols

# fit scaler on training data and save it
train_pre, scaler_example, sensor_cols = preprocess_df(train_df, cap_rul=125, sensor_cols=sensor_cols, fit_scaler=True)
joblib.dump(scaler_example, "scaler_fd003.joblib")
print("Fitted scaler saved to scaler_fd003.joblib")
print()

# 7 -- sequence generator
def generate_sequences(df, seq_len=50, sensor_cols=None, label_col="RUL_capped", step=1):
    Xs, ys = [], []
    units = sorted(df["unit"].unique())
    for u in units:
        tmp = df[df["unit"] == u].sort_values("cycle")
        data = tmp[sensor_cols].values
        labels = tmp[label_col].values
        n = data.shape[0]
        if n < seq_len:
            continue
        for start in range(0, n - seq_len + 1, step):
            end = start + seq_len
            Xs.append(data[start:end])
            ys.append(labels[end - 1])
    X = np.array(Xs)
    y = np.array(ys)
    return X, y

# 8 -- UNIT-WISE train / validation split (NO LEAKAGE)
SEQ_LEN = 50
VAL_RATIO = 0.15
RANDOM_STATE = 42

units = np.array(sorted(train_pre["unit"].unique()))
np.random.seed(RANDOM_STATE)
np.random.shuffle(units)

n_val = int(len(units) * VAL_RATIO)
val_units = units[:n_val]
train_units = units[n_val:]

print("Train units:", len(train_units))
print("Validation units:", len(val_units))

train_df_split = train_pre[train_pre["unit"].isin(train_units)]
val_df_split   = train_pre[train_pre["unit"].isin(val_units)]

# 9 -- generate sequences separately
X_train, y_train = generate_sequences(
    train_df_split, seq_len=SEQ_LEN, sensor_cols=sensor_cols
)

X_val, y_val = generate_sequences(
    val_df_split, seq_len=SEQ_LEN, sensor_cols=sensor_cols
)

print("X_train:", X_train.shape, "y_train:", y_train.shape)
print("X_val  :", X_val.shape, "y_val  :", y_val.shape)
print()

# 10 -- build GRU model
def build_gru_model(input_shape, units=128, dropout=0.3):
    model = Sequential([
        GRU(units, input_shape=input_shape, return_sequences=False),
        BatchNormalization(),
        Dropout(dropout),
        Dense(64, activation="relu"),
        Dense(1, activation="linear")
    ])
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3), loss="mse", metrics=["mae"])
    return model

model = build_gru_model(input_shape=(SEQ_LEN, len(sensor_cols)), units=128, dropout=0.3)
model.summary()

# 11 -- callbacks
checkpoint_path = "best_gru_fd003.h5"
callbacks = [
    ModelCheckpoint(checkpoint_path, monitor="val_loss", save_best_only=True, verbose=1),
    EarlyStopping(monitor="val_loss", patience=10, restore_best_weights=True, verbose=1)
]

# 12 -- train
history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=50,
    batch_size=64,
    callbacks=callbacks,
    verbose=1
)

# save final model (best is also saved by checkpoint)
model.save("final_gru_fd003.h5")
print("Final model saved to final_gru_fd003.h5 (best also in best_gru_fd003.h5)")
print()

# 13 -- plot training history
plt.figure(figsize=(8,4))
plt.plot(history.history["loss"], label="train_loss")
plt.plot(history.history["val_loss"], label="val_loss")
plt.xlabel("Epoch"); plt.ylabel("MSE"); plt.legend()
plt.title("Training History (FD003)")
plt.show()

# 14 -- validation evaluation (RMSE)
y_val_pred = model.predict(X_val).ravel()
rmse_val = math.sqrt(mean_squared_error(y_val, y_val_pred))
print(f"Validation RMSE (capped RUL units, FD003): {rmse_val:.4f}")
print()

# quick pred vs actual plot (random subset)
if len(y_val) >= 200:
    idx = np.random.choice(len(y_val), size=200, replace=False)
else:
    idx = np.arange(len(y_val))

plt.figure(figsize=(10,5))
plt.plot(y_val[idx], label="Actual RUL")
plt.plot(y_val_pred[idx], label="Predicted RUL")
plt.xlabel("Sample (Random Subset)"); plt.ylabel("RUL (Capped)")
plt.legend(); plt.title("Predicted vs Actual RUL (Validation Subset, FD003)")
plt.show()

# 15 -- Test evaluation (reads test_FD003.txt & RUL_FD003.txt)
test_file = os.path.join(DATA_DIR, "test_FD003.txt")
rul_file  = os.path.join(DATA_DIR, "RUL_FD003.txt")

if os.path.exists(test_file) and os.path.exists(rul_file):
    df_test_raw = read_cmapss(test_file)
    # ensure sensor_cols exists (we have it)
    print("Sensor cols used for test:", sensor_cols)
    # load scaler we saved earlier (or use scaler_example variable)
    if 'scaler_example' not in globals() or scaler_example is None:
        scaler_example = joblib.load("scaler_fd003.joblib")
        print("Loaded saved scaler from scaler_fd003.joblib")
    # scale test sensors
    df_test_scaled = df_test_raw.copy()
    df_test_scaled[sensor_cols] = scaler_example.transform(df_test_scaled[sensor_cols].values)
    # create X_test: last SEQ_LEN rows per unit (pad with first row if shorter)
    X_test = []
    test_units = sorted(df_test_scaled["unit"].unique())
    for u in test_units:
        tmp = df_test_scaled[df_test_scaled["unit"] == u].sort_values("cycle")
        arr = tmp[sensor_cols].values
        if arr.shape[0] >= SEQ_LEN:
            X_test.append(arr[-SEQ_LEN:])
        else:
            pad_n = SEQ_LEN - arr.shape[0]
            pad = np.vstack([arr[0]] * pad_n)   # pad with first row (you could use zeros)
            X_test.append(np.vstack([pad, arr]))
    X_test = np.array(X_test)
    # load true RULs and cap
    y_test_true = pd.read_csv(rul_file, sep=r"\s+", header=None).values.ravel()
    CAP_RUL = 125
    y_test_true = np.clip(y_test_true, None, CAP_RUL)
    # predict
    y_test_pred = model.predict(X_test).ravel()
    rmse_test = math.sqrt(mean_squared_error(y_test_true, y_test_pred))
    mae_test = np.mean(np.abs(y_test_true - y_test_pred))
    print(f"Test RMSE FD003 (capped to {CAP_RUL}): {rmse_test:.4f}")
    print(f"Test MAE  FD003 (capped to {CAP_RUL}): {mae_test:.4f}")
    # scatter
    plt.figure(figsize=(8,4))
    plt.scatter(y_test_true, y_test_pred, alpha=0.6)
    plt.plot([0, CAP_RUL], [0, CAP_RUL], "k--", linewidth=1)
    plt.xlim(0, CAP_RUL); plt.ylim(0, CAP_RUL)
    plt.xlabel("True RUL"); plt.ylabel("Predicted RUL"); plt.title("True vs Predicted (Test Set FD003)")
    plt.show()
else:
    print("Test or RUL file missing; skipping test evaluation.")

# 16 -- Table of Engine / True / Pred / Difference
if 'y_test_true' in globals() and 'y_test_pred' in globals():
    table_df = pd.DataFrame({
        "Engine": np.arange(1, len(y_test_true) + 1),
        "True RUL": y_test_true,
        "Predicted RUL": y_test_pred,
        "Error (Pred - True)": y_test_pred - y_test_true,
        "Absolute Error": np.abs(y_test_pred - y_test_true)
    })
    
    # Add percentage error
    table_df["Percent Error"] = np.where(
        table_df["True RUL"] > 0,
        100 * table_df["Error (Pred - True)"] / table_df["True RUL"],
        0
    )
    
    pd.set_option("display.max_rows", 200)
    pd.set_option("display.float_format", "{:.2f}".format)
    print("\nPred vs True table (first 50 rows):")
    print(table_df.head(50).to_string(index=False))
    print("\nFull table shape:", table_df.shape)
else:
    print("No test predictions available to build table.")

# 17 -- Maintenance Alerts using thresholds
WARNING_THRESHOLD = 30
CRITICAL_THRESHOLD = 10

def get_alert(rul_v):
    if rul_v <= CRITICAL_THRESHOLD:
        return "🔴 CRITICAL"
    elif rul_v <= WARNING_THRESHOLD:
        return "🟡 WARNING"
    else:
        return "🟢 NORMAL"

if 'y_test_pred' in globals():
    print("\n" + "="*70)
    print("MAINTENANCE ALERTS (First 20 Test Units, FD003)")
    print("="*70)
    
    for i in range(min(20, len(test_units))):
        alert = get_alert(y_test_pred[i])
        print(
            f"Engine {test_units[i]:3d}: "
            f"True RUL = {y_test_true[i]:6.1f}, "
            f"Pred RUL = {y_test_pred[i]:6.1f}, "
            f"Status: {alert}"
        )
    
    # Summary of alerts
    alerts = [get_alert(p) for p in y_test_pred]
    critical_count = sum(1 for a in alerts if "CRITICAL" in a)
    warning_count = sum(1 for a in alerts if "WARNING" in a)
    normal_count = sum(1 for a in alerts if "NORMAL" in a)
    
    print("\n" + "-"*70)
    print(f"Alert Summary (Total {len(test_units)} engines):")
    print(f"  🔴 CRITICAL: {critical_count} engines ({100*critical_count/len(test_units):.1f}%)")
    print(f"  🟡 WARNING:  {warning_count} engines ({100*warning_count/len(test_units):.1f}%)")
    print(f"  🟢 NORMAL:   {normal_count} engines ({100*normal_count/len(test_units):.1f}%)")
    
else:
    print("No test predictions to compute alerts.")

print("\n" + "="*70)
print("FD003 GRU PIPELINE COMPLETE")
print("="*70)
print("\nModel Performance Summary:")
print(f"  Validation RMSE: {rmse_val:.4f}")
if 'rmse_test' in globals():
    print(f"  Test RMSE:       {rmse_test:.4f}")
    print(f"  Test MAE:        {mae_test:.4f}")
print(f"  Sequence Length: {SEQ_LEN}")
print(f"  Features Used:   {len(sensor_cols)}")
print("="*70)
```
