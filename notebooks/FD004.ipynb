```python

# 1 -- Environment & imports
import os
import glob
import math
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense, Dropout, BatchNormalization, Bidirectional
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau

# 2 -- data directory & quick file listing
DATA_DIR = "/home/tejasri/Desktop/springboard"
txt_files = sorted(glob.glob(os.path.join(DATA_DIR, "*.txt")))
print("Files found:", txt_files)
print()

# 3 -- helper to read CMAPSS formatted files
def read_cmapss(file_path):
    col_names = ["unit", "cycle", "op_setting_1", "op_setting_2", "op_setting_3"]
    for i in range(1, 22):
        col_names.append(f"s{i}")
    df = pd.read_csv(file_path, sep=r"\s+", header=None, names=col_names)
    return df

# 4 -- add RUL to training dataframe
def add_rul_train(df):
    df = df.copy()
    max_cycle = df.groupby("unit")["cycle"].max().reset_index()
    max_cycle.columns = ["unit", "max_cycle"]
    df = df.merge(max_cycle, on="unit", how="left")
    df["RUL"] = df["max_cycle"] - df["cycle"]
    df.drop(columns=["max_cycle"], inplace=True)
    return df

# 5 -- locate train file for FD004 and read
train_fp = os.path.join(DATA_DIR, "train_FD004.txt")
if not os.path.exists(train_fp):
    raise FileNotFoundError(f"train_FD004.txt not found in {DATA_DIR}")

train_df = read_cmapss(train_fp)
train_df = add_rul_train(train_df)
print("Train df shape:", train_df.shape)
print("Units:", train_df["unit"].nunique())
print("Cycles per unit (min/max):", train_df.groupby("unit")["cycle"].max().min(), 
      "/", train_df.groupby("unit")["cycle"].max().max())
print()

# 6 -- Feature engineering: remove constant/low-variance sensors
def remove_constant_features(df, threshold=0.01):
    """Remove features with very low variance"""
    sensor_cols = [c for c in df.columns if c.startswith("s")]
    variances = df[sensor_cols].var()
    low_var = variances[variances < threshold].index.tolist()
    if low_var:
        print(f"Removing low-variance sensors: {low_var}")
        df = df.drop(columns=low_var)
    return df

# Remove constant sensors
train_df = remove_constant_features(train_df, threshold=0.01)

# detect sensor columns after removal
sensor_cols = [c for c in train_df.columns if c.startswith("s") or c.startswith("op_setting")]
print(f"Using {len(sensor_cols)} features:", sensor_cols)
print()

# 7 -- preprocessing function with MinMaxScaler for better GRU performance
def preprocess_df(df, cap_rul=125, sensor_cols=None, scaler=None, fit_scaler=True):
    df = df.copy()
    if sensor_cols is None:
        sensor_cols = [c for c in df.columns if c.startswith("s") or c.startswith("op_setting")]
    
    # Cap RUL
    if "RUL" in df.columns:
        df["RUL_capped"] = df["RUL"].clip(upper=cap_rul)
    
    # Scale sensors
    if fit_scaler:
        # Using MinMaxScaler for FD004 as it often works better with complex operating conditions
        scaler = MinMaxScaler(feature_range=(-1, 1))
        scaler.fit(df[sensor_cols])
    df[sensor_cols] = scaler.transform(df[sensor_cols])
    
    return df, scaler, sensor_cols

# fit scaler on training data and save it
train_pre, scaler_fd004, sensor_cols = preprocess_df(
    train_df, cap_rul=125, sensor_cols=sensor_cols, fit_scaler=True
)
joblib.dump(scaler_fd004, "scaler_fd004.joblib")
print("Fitted scaler saved to scaler_fd004.joblib")
print()

# 8 -- sequence generator with overlapping windows
def generate_sequences(df, seq_len=30, sensor_cols=None, label_col="RUL_capped", step=1):
    """Generate sequences with specified step size"""
    Xs, ys = [], []
    units = sorted(df["unit"].unique())
    
    for u in units:
        tmp = df[df["unit"] == u].sort_values("cycle")
        data = tmp[sensor_cols].values
        labels = tmp[label_col].values
        n = data.shape[0]
        
        if n < seq_len:
            # For short sequences, pad with the first row
            pad_n = seq_len - n
            pad = np.tile(data[0], (pad_n, 1))
            data = np.vstack([pad, data])
            labels = np.concatenate([np.full(pad_n, labels[0]), labels])
            n = seq_len
        
        for start in range(0, n - seq_len + 1, step):
            end = start + seq_len
            Xs.append(data[start:end])
            ys.append(labels[end - 1])
    
    X = np.array(Xs)
    y = np.array(ys)
    return X, y

# 9 -- UNIT-WISE train / validation split (NO LEAKAGE)
SEQ_LEN = 30  # Shorter sequence length for FD004 (more varied conditions)
VAL_RATIO = 0.15
RANDOM_STATE = 42

units = np.array(sorted(train_pre["unit"].unique()))
np.random.seed(RANDOM_STATE)
np.random.shuffle(units)

n_val = int(len(units) * VAL_RATIO)
val_units = units[:n_val]
train_units = units[n_val:]

print("Train units:", len(train_units))
print("Validation units:", len(val_units))

train_df_split = train_pre[train_pre["unit"].isin(train_units)]
val_df_split   = train_pre[train_pre["unit"].isin(val_units)]

# 10 -- generate sequences with smaller step for more training data
X_train, y_train = generate_sequences(
    train_df_split, seq_len=SEQ_LEN, sensor_cols=sensor_cols, step=1
)

X_val, y_val = generate_sequences(
    val_df_split, seq_len=SEQ_LEN, sensor_cols=sensor_cols, step=1
)

print("X_train:", X_train.shape, "y_train:", y_train.shape)
print("X_val  :", X_val.shape, "y_val  :", y_val.shape)
print("Mean RUL (train):", y_train.mean(), "Mean RUL (val):", y_val.mean())
print()

# 11 -- build enhanced GRU model for FD004
def build_gru_model(input_shape, units=128, dropout=0.3, use_bidirectional=False):
    """
    Build GRU model optimized for FD004
    - Deeper architecture for complex patterns
    - Optional bidirectional processing
    """
    model = Sequential()
    
    if use_bidirectional:
        model.add(Bidirectional(GRU(units, return_sequences=True), input_shape=input_shape))
    else:
        model.add(GRU(units, return_sequences=True, input_shape=input_shape))
    
    model.add(BatchNormalization())
    model.add(Dropout(dropout))
    
    model.add(GRU(units//2, return_sequences=False))
    model.add(BatchNormalization())
    model.add(Dropout(dropout))
    
    model.add(Dense(64, activation="relu"))
    model.add(Dropout(dropout/2))
    model.add(Dense(32, activation="relu"))
    model.add(Dense(1, activation="linear"))
    
    # Using Adam with slightly lower learning rate for stability
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=5e-4),
        loss="mse",
        metrics=["mae"]
    )
    return model

model = build_gru_model(
    input_shape=(SEQ_LEN, len(sensor_cols)),
    units=128,
    dropout=0.3,
    use_bidirectional=False
)
model.summary()

# 12 -- callbacks with learning rate reduction
checkpoint_path = "best_gru_fd004.h5"
callbacks = [
    ModelCheckpoint(
        checkpoint_path,
        monitor="val_loss",
        save_best_only=True,
        verbose=1
    ),
    EarlyStopping(
        monitor="val_loss",
        patience=15,
        restore_best_weights=True,
        verbose=1
    ),
    ReduceLROnPlateau(
        monitor="val_loss",
        factor=0.5,
        patience=5,
        min_lr=1e-6,
        verbose=1
    )
]

# 13 -- train with larger batch size for stability
print("Starting training...")
history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=80,
    batch_size=128,
    callbacks=callbacks,
    verbose=1
)

# save final model
model.save("final_gru_fd004.h5")
print("\nFinal model saved to final_gru_fd004.h5 (best also in best_gru_fd004.h5)")
print()

# 14 -- plot training history
fig, axes = plt.subplots(1, 2, figsize=(14, 4))

axes[0].plot(history.history["loss"], label="train_loss")
axes[0].plot(history.history["val_loss"], label="val_loss")
axes[0].set_xlabel("Epoch")
axes[0].set_ylabel("MSE")
axes[0].legend()
axes[0].set_title("Training History - Loss (FD004)")
axes[0].grid(True, alpha=0.3)

axes[1].plot(history.history["mae"], label="train_mae")
axes[1].plot(history.history["val_mae"], label="val_mae")
axes[1].set_xlabel("Epoch")
axes[1].set_ylabel("MAE")
axes[1].legend()
axes[1].set_title("Training History - MAE (FD004)")
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# 15 -- validation evaluation (RMSE)
y_val_pred = model.predict(X_val, verbose=0).ravel()
rmse_val = math.sqrt(mean_squared_error(y_val, y_val_pred))
mae_val = np.mean(np.abs(y_val - y_val_pred))

print(f"Validation RMSE (capped RUL, FD004): {rmse_val:.4f}")
print(f"Validation MAE  (capped RUL, FD004): {mae_val:.4f}")
print()

# Prediction distribution analysis
print("Validation Predictions Analysis:")
print(f"  True RUL range: [{y_val.min():.1f}, {y_val.max():.1f}]")
print(f"  Pred RUL range: [{y_val_pred.min():.1f}, {y_val_pred.max():.1f}]")
print()

# quick pred vs actual plot (random subset)
n_samples = min(300, len(y_val))
idx = np.random.choice(len(y_val), size=n_samples, replace=False)
idx = np.sort(idx)

plt.figure(figsize=(12, 5))
plt.plot(y_val[idx], label="Actual RUL", alpha=0.7, linewidth=2)
plt.plot(y_val_pred[idx], label="Predicted RUL", alpha=0.7, linewidth=2)
plt.xlabel("Sample Index")
plt.ylabel("RUL (Capped)")
plt.legend()
plt.title(f"Predicted vs Actual RUL (Validation, FD004) - RMSE: {rmse_val:.2f}")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# 16 -- Test evaluation
test_file = os.path.join(DATA_DIR, "test_FD004.txt")
rul_file  = os.path.join(DATA_DIR, "RUL_FD004.txt")

if os.path.exists(test_file) and os.path.exists(rul_file):
    print("Loading test data...")
    df_test_raw = read_cmapss(test_file)
    
    # Remove the same sensors as in training
    df_test_raw = df_test_raw[[c for c in df_test_raw.columns if c in train_df.columns or c == 'unit' or c == 'cycle']]
    
    print("Test df shape:", df_test_raw.shape)
    print("Test units:", df_test_raw["unit"].nunique())
    
    # Scale test sensors using the saved scaler
    df_test_scaled = df_test_raw.copy()
    df_test_scaled[sensor_cols] = scaler_fd004.transform(df_test_scaled[sensor_cols].values)
    
    # Create X_test: last SEQ_LEN rows per unit (pad if shorter)
    X_test = []
    test_units = sorted(df_test_scaled["unit"].unique())
    
    for u in test_units:
        tmp = df_test_scaled[df_test_scaled["unit"] == u].sort_values("cycle")
        arr = tmp[sensor_cols].values
        
        if arr.shape[0] >= SEQ_LEN:
            X_test.append(arr[-SEQ_LEN:])
        else:
            # Pad with first row repeated
            pad_n = SEQ_LEN - arr.shape[0]
            pad = np.tile(arr[0], (pad_n, 1))
            X_test.append(np.vstack([pad, arr]))
    
    X_test = np.array(X_test)
    print("X_test shape:", X_test.shape)
    
    # Load true RULs and cap
    y_test_true = pd.read_csv(rul_file, sep=r"\s+", header=None).values.ravel()
    CAP_RUL = 125
    y_test_true = np.clip(y_test_true, None, CAP_RUL)
    
    print("y_test_true shape:", y_test_true.shape)
    print(f"True RUL range: [{y_test_true.min():.1f}, {y_test_true.max():.1f}]")
    
    # Predict
    print("\nGenerating test predictions...")
    y_test_pred = model.predict(X_test, verbose=0).ravel()
    
    # Compute metrics
    rmse_test = math.sqrt(mean_squared_error(y_test_true, y_test_pred))
    mae_test = np.mean(np.abs(y_test_true - y_test_pred))
    
    print(f"\n{'='*60}")
    print(f"Test RMSE (FD004, capped to {CAP_RUL}): {rmse_test:.4f}")
    print(f"Test MAE  (FD004, capped to {CAP_RUL}): {mae_test:.4f}")
    print(f"{'='*60}\n")
    
    # Scatter plot
    plt.figure(figsize=(10, 8))
    plt.scatter(y_test_true, y_test_pred, alpha=0.5, s=30)
    plt.plot([0, CAP_RUL], [0, CAP_RUL], "r--", linewidth=2, label="Perfect Prediction")
    plt.xlim(0, CAP_RUL)
    plt.ylim(0, CAP_RUL)
    plt.xlabel("True RUL", fontsize=12)
    plt.ylabel("Predicted RUL", fontsize=12)
    plt.title(f"True vs Predicted RUL (Test Set, FD004)\nRMSE: {rmse_test:.2f}, MAE: {mae_test:.2f}", fontsize=13)
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    # Error distribution
    errors = y_test_pred - y_test_true
    plt.figure(figsize=(10, 5))
    plt.hist(errors, bins=50, edgecolor='black', alpha=0.7)
    plt.xlabel("Prediction Error (Pred - True)")
    plt.ylabel("Frequency")
    plt.title(f"Error Distribution (Test Set, FD004)\nMean Error: {errors.mean():.2f}, Std: {errors.std():.2f}")
    plt.axvline(0, color='red', linestyle='--', linewidth=2, label='Zero Error')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    
else:
    print("Test or RUL file missing; skipping test evaluation.")

# 17 -- Table of Engine / True / Pred / Difference
if 'y_test_true' in globals() and 'y_test_pred' in globals():
    table_df = pd.DataFrame({
        "Engine": np.arange(1, len(y_test_true) + 1),
        "True RUL": y_test_true,
        "Predicted RUL": y_test_pred,
        "Error (Pred - True)": y_test_pred - y_test_true,
        "Absolute Error": np.abs(y_test_pred - y_test_true)
    })
    
    # Add percentage error
    table_df["Percent Error"] = np.where(
        table_df["True RUL"] > 0,
        100 * table_df["Error (Pred - True)"] / table_df["True RUL"],
        0
    )
    
    pd.set_option("display.max_rows", 200)
    pd.set_option("display.float_format", "{:.2f}".format)
    
    print("\nPrediction Results (first 30 engines):")
    print(table_df.head(30).to_string(index=False))
    
    print(f"\nSummary Statistics:")
    print(f"  Mean Absolute Error: {table_df['Absolute Error'].mean():.2f}")
    print(f"  Median Absolute Error: {table_df['Absolute Error'].median():.2f}")
    print(f"  Max Error: {table_df['Absolute Error'].max():.2f}")
    print(f"  Min Error: {table_df['Absolute Error'].min():.2f}")
    
else:
    print("No test predictions available to build table.")

# 18 -- Maintenance Alerts using thresholds
WARNING_THRESHOLD = 30
CRITICAL_THRESHOLD = 10

def get_alert(rul_v):
    if rul_v <= CRITICAL_THRESHOLD:
        return "🔴 CRITICAL"
    elif rul_v <= WARNING_THRESHOLD:
        return "🟡 WARNING"
    else:
        return "🟢 NORMAL"

if 'y_test_pred' in globals():
    print("\n" + "="*70)
    print("MAINTENANCE ALERTS (First 20 Test Units, FD004)")
    print("="*70)
    
    for i in range(min(20, len(test_units))):
        alert = get_alert(y_test_pred[i])
        print(
            f"Engine {test_units[i]:3d}: "
            f"True RUL = {y_test_true[i]:6.1f}, "
            f"Pred RUL = {y_test_pred[i]:6.1f}, "
            f"Status: {alert}"
        )
    
    # Summary of alerts
    alerts = [get_alert(p) for p in y_test_pred]
    critical_count = sum(1 for a in alerts if "CRITICAL" in a)
    warning_count = sum(1 for a in alerts if "WARNING" in a)
    normal_count = sum(1 for a in alerts if "NORMAL" in a)
    
    print("\n" + "-"*70)
    print(f"Alert Summary (Total {len(test_units)} engines):")
    print(f"  🔴 CRITICAL: {critical_count} engines ({100*critical_count/len(test_units):.1f}%)")
    print(f"  🟡 WARNING:  {warning_count} engines ({100*warning_count/len(test_units):.1f}%)")
    print(f"  🟢 NORMAL:   {normal_count} engines ({100*normal_count/len(test_units):.1f}%)")
    
else:
    print("No test predictions to compute alerts.")

print("\n" + "="*70)
print("FD004 GRU PIPELINE COMPLETE")
print("="*70)
print("\nModel Performance Summary:")
print(f"  Validation RMSE: {rmse_val:.4f}")
if 'rmse_test' in globals():
    print(f"  Test RMSE:       {rmse_test:.4f}")
print(f"  Sequence Length: {SEQ_LEN}")
print(f"  Features Used:   {len(sensor_cols)}")
print("="*70)
```
