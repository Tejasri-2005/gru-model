```python

# 1 -- Environment & Imports
# =================================================================
import os
import glob
import math
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping

# =================================================================
# 2 -- Configuration & Data Directory
# =================================================================
DATA_DIR = "/home/tejasri/Desktop/springboard"
DATASET = "FD001"
SEQ_LEN = 50
CAP_RUL = 125
VAL_RATIO = 0.15
RANDOM_STATE = 42

txt_files = sorted(glob.glob(os.path.join(DATA_DIR, "*.txt")))
print("Files found:", txt_files)
print()

# =================================================================
# 3 -- Helper Function: Read CMAPSS Files
# =================================================================
def read_cmapss(file_path):
    """Read CMAPSS dataset files with proper column names."""
    col_names = ["unit", "cycle", "op_setting_1", "op_setting_2", "op_setting_3"]
    for i in range(1, 22):
        col_names.append(f"s{i}")
    df = pd.read_csv(file_path, sep=r"\s+", header=None, names=col_names)
    return df

# =================================================================
# 4 -- Helper Function: Add RUL to Training Data
# =================================================================
def add_rul_train(df):
    """Calculate Remaining Useful Life (RUL) for training data."""
    df = df.copy()
    max_cycle = df.groupby("unit")["cycle"].max().reset_index()
    max_cycle.columns = ["unit", "max_cycle"]
    df = df.merge(max_cycle, on="unit", how="left")
    df["RUL"] = df["max_cycle"] - df["cycle"]
    df.drop(columns=["max_cycle"], inplace=True)
    return df

# =================================================================
# 5 -- Load Training Data
# =================================================================
train_fp = os.path.join(DATA_DIR, f"train_{DATASET}.txt")
if not os.path.exists(train_fp):
    raise FileNotFoundError(f"train_{DATASET}.txt not found in {DATA_DIR}")

train_df = read_cmapss(train_fp)
train_df = add_rul_train(train_df)
print(f"Train df shape: {train_df.shape}")
print(f"Number of engines: {train_df['unit'].nunique()}")
print()

# Detect sensor columns
sensor_cols = [c for c in train_df.columns 
               if c.startswith("s") or c.startswith("op_setting")]
print(f"Total sensor columns: {len(sensor_cols)}")
print(f"Sensor cols (first 8): {sensor_cols[:8]}")
print()

# =================================================================
# 6 -- Preprocessing Function
# =================================================================
def preprocess_df(df, cap_rul=125, sensor_cols=None, scaler=None, fit_scaler=True):
    """Preprocess data: cap RUL and scale features."""
    df = df.copy()
    if sensor_cols is None:
        sensor_cols = [c for c in df.columns 
                      if c.startswith("s") or c.startswith("op_setting")]
    
    # Cap RUL at specified threshold
    df["RUL_capped"] = df["RUL"].clip(upper=cap_rul)
    
    # Fit or apply scaler
    if fit_scaler:
        scaler = StandardScaler()
        scaler.fit(df[sensor_cols])
    
    df[sensor_cols] = scaler.transform(df[sensor_cols])
    
    return df, scaler, sensor_cols

# Fit scaler on training data and save
train_pre, scaler_example, sensor_cols = preprocess_df(
    train_df, cap_rul=CAP_RUL, sensor_cols=sensor_cols, fit_scaler=True
)
joblib.dump(scaler_example, f"scaler_{DATASET}.joblib")
print(f"Fitted scaler saved to scaler_{DATASET}.joblib")
print()

# =================================================================
# 7 -- Sequence Generator Function
# =================================================================
def generate_sequences(df, seq_len=50, sensor_cols=None, label_col="RUL_capped", step=1):
    """Generate sequences from time-series data."""
    Xs, ys = [], []
    units = sorted(df["unit"].unique())
    
    for u in units:
        tmp = df[df["unit"] == u].sort_values("cycle")
        data = tmp[sensor_cols].values
        labels = tmp[label_col].values
        n = data.shape[0]
        
        if n < seq_len:
            continue
        
        for start in range(0, n - seq_len + 1, step):
            end = start + seq_len
            Xs.append(data[start:end])
            ys.append(labels[end - 1])  # Label at last time step
    
    X = np.array(Xs)
    y = np.array(ys)
    
    return X, y

# =================================================================
# 8 -- Unit-Wise Train/Validation Split 
# =================================================================
units = np.array(sorted(train_pre["unit"].unique()))
np.random.seed(RANDOM_STATE)
np.random.shuffle(units)

n_val = int(len(units) * VAL_RATIO)
val_units = units[:n_val]
train_units = units[n_val:]

print(f"Train units: {len(train_units)}")
print(f"Validation units: {len(val_units)}")
print()

train_df_split = train_pre[train_pre["unit"].isin(train_units)]
val_df_split = train_pre[train_pre["unit"].isin(val_units)]

# =================================================================
# 9 -- Generate Sequences Separately for Train and Validation
# =================================================================
X_train, y_train = generate_sequences(
    train_df_split, seq_len=SEQ_LEN, sensor_cols=sensor_cols
)

X_val, y_val = generate_sequences(
    val_df_split, seq_len=SEQ_LEN, sensor_cols=sensor_cols
)

print(f"X_train: {X_train.shape}, y_train: {y_train.shape}")
print(f"X_val:   {X_val.shape}, y_val:   {y_val.shape}")
print()

# =================================================================
# 10 -- Build GRU Model
# =================================================================
def build_gru_model(input_shape, units=128, dropout=0.3):
    """Build GRU model for RUL prediction."""
    model = Sequential([
        GRU(units, input_shape=input_shape, return_sequences=False),
        BatchNormalization(),
        Dropout(dropout),
        Dense(64, activation="relu"),
        Dense(1, activation="linear")
    ])
    
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
        loss="mse",
        metrics=["mae"]
    )
    
    return model

model = build_gru_model(
    input_shape=(SEQ_LEN, len(sensor_cols)),
    units=128,
    dropout=0.3
)

model.summary()

# =================================================================
# 11 -- Training Callbacks
# =================================================================
checkpoint_path = f"best_gru_{DATASET}.h5"
callbacks = [
    ModelCheckpoint(
        checkpoint_path,
        monitor="val_loss",
        save_best_only=True,
        verbose=1
    ),
    EarlyStopping(
        monitor="val_loss",
        patience=10,
        restore_best_weights=True,
        verbose=1
    )
]

# =================================================================
# 12 -- Train the Model
# =================================================================
history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=50,
    batch_size=64,
    callbacks=callbacks,
    verbose=1
)

# Save final model
model.save(f"final_gru_{DATASET}.h5")
print(f"\nFinal model saved to final_gru_{DATASET}.h5")
print(f"Best model saved to {checkpoint_path}")
print()

# =================================================================
# 13 -- Plot Training History
# =================================================================
plt.figure(figsize=(10, 5))
plt.plot(history.history["loss"], label="Train Loss", linewidth=2)
plt.plot(history.history["val_loss"], label="Val Loss", linewidth=2)
plt.xlabel("Epoch", fontsize=12)
plt.ylabel("MSE Loss", fontsize=12)
plt.legend(fontsize=11)
plt.title(f"Training History ({DATASET})", fontsize=14)
plt.grid(True, alpha=0.3)
plt.show()

# =================================================================
# 14 -- Validation Evaluation (RMSE)
# =================================================================
y_val_pred = model.predict(X_val).ravel()
rmse_val = math.sqrt(mean_squared_error(y_val, y_val_pred))
mae_val = np.mean(np.abs(y_val - y_val_pred))

print(f"Validation RMSE (capped RUL, {DATASET}): {rmse_val:.4f}")
print(f"Validation MAE  (capped RUL, {DATASET}): {mae_val:.4f}")
print()

# Plot predicted vs actual 
if len(y_val) >= 200:
    idx = np.random.choice(len(y_val), size=200, replace=False)
else:
    idx = np.arange(len(y_val))

plt.figure(figsize=(12, 6))
plt.plot(y_val[idx], label="Actual RUL", linewidth=2, alpha=0.7)
plt.plot(y_val_pred[idx], label="Predicted RUL", linewidth=2, alpha=0.7)
plt.xlabel("Sample (Random Subset)", fontsize=12)
plt.ylabel("RUL (Capped)", fontsize=12)
plt.legend(fontsize=11)
plt.title(f"Predicted vs Actual RUL (Validation, {DATASET})", fontsize=14)
plt.grid(True, alpha=0.3)
plt.show()

# =================================================================
# 15 -- Test Set Evaluation
# =================================================================
test_file = os.path.join(DATA_DIR, f"test_{DATASET}.txt")
rul_file = os.path.join(DATA_DIR, f"RUL_{DATASET}.txt")

if os.path.exists(test_file) and os.path.exists(rul_file):
    print(f"Loading test data for {DATASET}...")
    
    # Load test data
    df_test_raw = read_cmapss(test_file)
    print(f"Test data shape: {df_test_raw.shape}")
    
    # Load scaler
    if 'scaler_example' not in globals() or scaler_example is None:
        scaler_example = joblib.load(f"scaler_{DATASET}.joblib")
        print(f"Loaded saved scaler from scaler_{DATASET}.joblib")
    
    # Scale test sensors
    df_test_scaled = df_test_raw.copy()
    df_test_scaled[sensor_cols] = scaler_example.transform(
        df_test_scaled[sensor_cols].values
    )
    
    # Create X_test: last SEQ_LEN rows per unit (pad if shorter)
    X_test = []
    test_units = sorted(df_test_scaled["unit"].unique())
    
    for u in test_units:
        tmp = df_test_scaled[df_test_scaled["unit"] == u].sort_values("cycle")
        arr = tmp[sensor_cols].values
        
        if arr.shape[0] >= SEQ_LEN:
            X_test.append(arr[-SEQ_LEN:])
        else:
            # Pad with first row if shorter than SEQ_LEN
            pad_n = SEQ_LEN - arr.shape[0]
            pad = np.vstack([arr[0]] * pad_n)
            X_test.append(np.vstack([pad, arr]))
    
    X_test = np.array(X_test)
    
    # Load true RULs and cap
    y_test_true = pd.read_csv(rul_file, sep=r"\s+", header=None).values.ravel()
    y_test_true = np.clip(y_test_true, None, CAP_RUL)
    
    # Predict
    y_test_pred = model.predict(X_test).ravel()
    
    # Calculate metrics
    rmse_test = math.sqrt(mean_squared_error(y_test_true, y_test_pred))
    mae_test = np.mean(np.abs(y_test_true - y_test_pred))
    
    print(f"\nTest RMSE {DATASET} (capped to {CAP_RUL}): {rmse_test:.4f}")
    print(f"Test MAE  {DATASET} (capped to {CAP_RUL}): {mae_test:.4f}")
    print()
    
    # Scatter plot: True vs Predicted
    plt.figure(figsize=(8, 8))
    plt.scatter(y_test_true, y_test_pred, alpha=0.6, s=30)
    plt.plot([0, CAP_RUL], [0, CAP_RUL], "k--", linewidth=2, label="Perfect Prediction")
    plt.xlim(0, CAP_RUL)
    plt.ylim(0, CAP_RUL)
    plt.xlabel("True RUL", fontsize=12)
    plt.ylabel("Predicted RUL", fontsize=12)
    plt.title(f"True vs Predicted RUL (Test Set, {DATASET})", fontsize=14)
    plt.legend(fontsize=11)
    plt.grid(True, alpha=0.3)
    plt.show()
    
else:
    print(f"Test or RUL file missing for {DATASET}; skipping test evaluation.")

# =================================================================
# 16 -- Prediction Table
# =================================================================
if 'y_test_true' in globals() and 'y_test_pred' in globals():
    table_df = pd.DataFrame({
        "Engine": np.arange(1, len(y_test_true) + 1),
        "True RUL": y_test_true,
        "Predicted RUL": y_test_pred,
        "Error (Pred - True)": y_test_pred - y_test_true,
        "Absolute Error": np.abs(y_test_pred - y_test_true)
    })
    
    # Add percentage error
    table_df["Percent Error"] = np.where(
        table_df["True RUL"] > 0,
        100 * table_df["Error (Pred - True)"] / table_df["True RUL"],
        0
    )
    
    pd.set_option("display.max_rows", 200)
    pd.set_option("display.float_format", "{:.2f}".format)
    
    print("\n" + "="*80)
    print("PREDICTION TABLE (First 50 Engines)")
    print("="*80)
    print(table_df.head(50).to_string(index=False))
    print(f"\nFull table shape: {table_df.shape}")
    print()
    
else:
    print("No test predictions available to build table.")

# =================================================================
# 17 -- Maintenance Alerts
# =================================================================
WARNING_THRESHOLD = 30
CRITICAL_THRESHOLD = 10

def get_alert(rul_v):
    """Determine maintenance alert level based on predicted RUL."""
    if rul_v <= CRITICAL_THRESHOLD:
        return "🔴 CRITICAL"
    elif rul_v <= WARNING_THRESHOLD:
        return "🟡 WARNING"
    else:
        return "🟢 NORMAL"

if 'y_test_pred' in globals():
    print("="*80)
    print(f"MAINTENANCE ALERTS (First 20 Test Units, {DATASET})")
    print("="*80)
    
    for i in range(min(20, len(test_units))):
        alert = get_alert(y_test_pred[i])
        print(
            f"Engine {test_units[i]:3d}: "
            f"True RUL = {y_test_true[i]:6.1f}, "
            f"Pred RUL = {y_test_pred[i]:6.1f}, "
            f"Status: {alert}"
        )
    
    # Summary of alerts
    alerts = [get_alert(p) for p in y_test_pred]
    critical_count = sum(1 for a in alerts if "CRITICAL" in a)
    warning_count = sum(1 for a in alerts if "WARNING" in a)
    normal_count = sum(1 for a in alerts if "NORMAL" in a)
    
    print("\n" + "-"*80)
    print(f"Alert Summary (Total {len(test_units)} engines):")
    print(f"  🔴 CRITICAL: {critical_count:3d} engines ({100*critical_count/len(test_units):5.1f}%)")
    print(f"  🟡 WARNING:  {warning_count:3d} engines ({100*warning_count/len(test_units):5.1f}%)")
    print(f"  🟢 NORMAL:   {normal_count:3d} engines ({100*normal_count/len(test_units):5.1f}%)")
    
else:
    print("No test predictions to compute alerts.")

# =================================================================
# 18 -- Final Summary
# =================================================================
print("\n" + "="*80)
print(f"{DATASET} GRU PIPELINE COMPLETE")
print("="*80)
print("\nModel Performance Summary:")
print(f"  Validation RMSE: {rmse_val:.4f}")
print(f"  Validation MAE:  {mae_val:.4f}")
if 'rmse_test' in globals():
    print(f"  Test RMSE:       {rmse_test:.4f}")
    print(f"  Test MAE:        {mae_test:.4f}")
print(f"\nConfiguration:")
print(f"  Dataset:         {DATASET}")
print(f"  Sequence Length: {SEQ_LEN}")
print(f"  RUL Cap:         {CAP_RUL}")
print(f"  Features Used:   {len(sensor_cols)}")
print(f"  Train Engines:   {len(train_units)}")
print(f"  Val Engines:     {len(val_units)}")
if 'test_units' in globals():
    print(f"  Test Engines:    {len(test_units)}")
print("="*80)
```
